```python
import matplotlib.pyplot as plt
import jax.numpy as jnp
from jax import grad, vmap
```

```python
fig, ax = plt.subplots()

x = jnp.linspace(-2 * jnp.pi, 2 * jnp.pi, 1000)

y = jnp.cos
d1y = grad(y)
d2y = grad(grad(y))

ax.plot(x, y(x), 'k-', lw=4)
ax.plot(x, vmap(d1y)(x), 'b-')
ax.plot(x, -vmap(d2y)(x), 'w--')
```
